```python
import numpy as np
import gudhi
import json
import csv
import matplotlib
import matplotlib.pyplot as plot
import import_ipynb
import vineyards
from vineyards import *
```

```python
visual=True #dispalys PH[1] for each step if True



#points = np.loadtxt('simulation/1.txt')  
points = np.loadtxt('otto/otto_1.txt')  

#points=np.delete(points, 2, axis=1) #only x y coordinates
```

```python
##---- First step----##
rips_complex = gudhi.RipsComplex(points=points)
simplex_tree = rips_complex.create_simplex_tree(max_dimension=2) 
flt=  simplex_tree.get_filtration(); 
flt = sorted(flt, key=lambda x: (len(x[0]), x[0])) #per vine implementation
flt_val = [val for _, val in flt] #list comprehension, skippa il primo e prende il secondo
flt_nodes=[first for first ,_ in  flt]
flt_time =flt_val
```

```python
##----same but in loop to collect every filtration value----##
```

```python

for step in range (2,26): 
    
    #points = np.loadtxt('simulation/'+str(step)+'.txt') 
    points = np.loadtxt('otto/otto_'+str(step)+'.txt')  
    points=np.delete(points, 2, axis=1) #only x y coordinates
    rips_complex = gudhi.RipsComplex(points=points)
    simplex_tree = rips_complex.create_simplex_tree(max_dimension=2) 
    flt=  simplex_tree.get_filtration(); 
    flt = sorted(flt, key=lambda x: (len(x[0]), x[0])) #per vine implementation
    flt_val = [val for _, val in flt] #list comprehension, skippa il primo e prende il secondo
    flt_time=np.column_stack((flt_time, flt_val));
    if visual==True:
        PH = simplex_tree.persistence()
        PHs = [[y for y in PH if y[0]==x] for x in [0,1]]
        gudhi.plot_persistence_diagram(PHs[1], legend=True)
        print(len(PHs[1]))    #check di quante classi ci sono 
```
