```python
import matplotlib.pyplot as plt
import importlib
import torch
import numpy as np
from tqdm import tqdm
import hypernetx as hnx
import tpot
```

```python
from pathlib import Path
import json
import re

# Define the folder containing the JSON files as a Path object (portable)
folder = Path("PH") / f"PH_{name}"

H_all = []  # List to store all loaded JSON data

# Function to extract a number from the filename (used for sorting)
def extract_number(filename):
    match = re.search(r'\d+', filename)
    return int(match.group()) if match else -1

# Get a sorted list of all JSON files in the folder, sorted by the extracted number
files_sorted = sorted(folder.glob('*.json'), key=lambda x: extract_number(x.name))

# Loop through each sorted file, open it, and load the JSON data
for file_path in files_sorted:
    with open(file_path, 'r') as f:
        H = json.load(f)
        H_all.append(H)

# Example of constructing Hypergraph objects from loaded data (commented)
# G_all = [hnx.Hypergraph({str(i): k for (i, k) in enumerate(g['representatives'])}) for g in H_all]
```

```python
# Creiamo un dizionario che mappa edge_uid a set di nodi
edge_dict = {}
for g in H_all:
    for i, nodes in enumerate(g['representatives']):
        edge_dict[str(i)] = set(map(str, nodes))

# Ora creiamo l'hypergraph da questo dict
#G_all = hnx.Hypergraph(edge_dict)
```

```python
G_all = [hnx.Hypergraph({str(i): set(map(str, nodes)) for i, nodes in enumerate(g['representatives'])}) for g in H_all]
```

```python


TOP_matchings = []
GEO_matchings = []
SOURCE = []
TARGET = []
VERT_matchings = []
h = 0.4

for i in range(len(Y_figure)-1):
    j = i+1
    # topological matching 
    df1 = topo_util.make_dataframe(H_all[i]["barcode"])
    df2 = topo_util.make_dataframe(H_all[j]["barcode"])
    p_spt = df1.iloc[:, 0:2].to_numpy()
    q_spt = df2.iloc[:, 0:2].to_numpy()
    p, q = pd.augmented_weights(p_spt, q_spt)
    C = pd.augmented_cost(p_spt, q_spt); C /= C.mean()
    pi = ot.emd(p, q, C)
    # TPOT matching 
    y0, w0, v0 = topo_util.process_hg(G_all[i], Y_figure[i].shape[0])
    y1, w1, v1 = topo_util.process_hg(G_all[j], Y_figure[j].shape[0])
    K0 = topo_util.get_kernel(Y_figure[i], h)
    L0 = topo_util.symmetric_laplacian(K0)
    K0 /= K0.sum(-1).reshape(-1, 1)
    K1 = topo_util.get_kernel(Y_figure[j], h)
    L1 = topo_util.symmetric_laplacian(K1)
    K1 /= K1.sum(-1).reshape(-1, 1)
    # impute generator hyperedges
    y0_imputed = topo_util.impute_laplacian(y0, L0)
    y1_imputed = topo_util.impute_laplacian(y1, L1)
    y0 = y0_imputed; y1 = y1_imputed;
    TARGET.append(y1)
    SOURCE.append(y0)
    # setup cost matrices
    C_v0 = sp.spatial.distance.cdist(Y_figure[i], Y_figure[i], metric = "euclidean"); C_v0 /= np.mean(C_v0)
    # C_v0 = np.exp(-C_v0)
    C_v1 = sp.spatial.distance.cdist(Y_figure[j], Y_figure[j], metric = "euclidean"); C_v1 /= np.mean(C_v1)
    # C_v1 = np.exp(-C_v1)
    C_pd = pd.augmented_cost(p_spt, q_spt); C_pd /= np.mean(C_pd)
    v0 = p
    v1 = q
    pi_s, pi_f, _, = tpot.TPOT(y0, y1, w0, w1, v0, v1,
                              C_v0, C_v1, C_pd, 
                              alpha = 0.5, beta = 10.0, print_iter = 50, iter = 100, 
                              eps_s = 0, eps_f = 0,
                              solver = "exact")
    # pi_f = ot.emd(v0, v1, pi_f.max() - pi_f)
    # pi_s = ot.emd(w0, w1, pi_s.max() - pi_s)
    t0, t1 = t, t
    P = (pi_s / pi_s.sum(axis=-1)[:, np.newaxis])
    Q = (pi_f[:-1, :-1] / pi_f[:-1, :-1].sum(axis=-1)[:, np.newaxis])
    Q_pd = pi[:-1, :-1] / pi[:-1, :-1].sum(axis=-1)[:, np.newaxis]
    Q[np.isnan(Q)] = 0
    Q_pd[np.isnan(Q_pd)] = 0
    TOP_matchings.append(Q_pd)
    GEO_matchings.append(Q)
    VERT_matchings.append(P)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of cycles to track
n_cycles_to_track = 5
what = [-i for i in range(1, n_cycles_to_track + 1)]

# Fixed colors manually chosen for each cycle
cycle_colors = ['red', 'green', 'blue', 'purple', 'orange']  # you can change as desired

# Include the last frame of TARGET
_ys = SOURCE + [TARGET[-1], ] 

# Create a figure with a specific size
fig = plt.figure(figsize=(10, 25))

for i in range(len(_ys)):
    y0 = _ys[i]
    A = Y_figure[i]

    max_c = y0.shape[1]  # number of cycles present in this frame

    for l, w in enumerate(what):
        if abs(w) > max_c:
            print(f"Frame {i}: cycle {w} no longer exists (only {max_c} cycles)")
            continue

        # Create a 3D subplot for each cycle
        ax = fig.add_subplot(len(_ys), n_cycles_to_track, i * n_cycles_to_track + l + 1, projection='3d')

        # Fixed color for each cycle
        color = cycle_colors[l % len(cycle_colors)]

        # Fixed marker size (optional, here same for all)
        marker_size = 20

        # Scatter plot for background points in light gray
        ax.scatter(
            A[:,1], A[:,0], A[:,2],
            s=marker_size,
            color='lightgray',
            edgecolor='k',
            linewidth=0.5,
            alpha=0.3,
            rasterized=True
        )
        
        # Scatter plot for the current cycle with size proportional to y0
        ax.scatter(
            A[:,1], A[:,0], A[:,2],
            s=2.5 + 25 * y0[:, w],
            edgecolor='k',
            linewidth=0.8,
            color=color,
            rasterized=True
        )

        # Hide axes
        ax.set_axis_off()
        # Optional: set the viewing angle
        # ax.view_init(elev=50, azim=20+i, roll=10)

    # Update 'what' using geometric matchings
    if i < len(GEO_matchings):
        Q = GEO_matchings[i]
        if Q.shape[0] > max(abs(w) for w in what):
            # Update indices of cycles
            what = np.argmax(Q, axis=1)[what]
            # Keep negatives by convention
            what = [-(Q.shape[1] - w) for w in what]
        else:
            print(f"Frame {i}: matching Q too small to update 'what'")

# Adjust layout to prevent overlaps
plt.tight_layout()
plt.show()
```

```python
##---VISUALIZE each cycles evolution by row

# Number of cycles to track
n_cycles_to_track = 4

# Fixed colors for each cycle
cycle_colors = ['red', 'green', 'blue', 'purple', 'orange']

# Indices of cycles to track (negative convention)
what = [-i for i in range(1, n_cycles_to_track + 1)]

# Time steps to visualize
steps_to_plot = [0, 5, 10, 15, 20]

# Create a figure with n_cycles_to_track rows and len(steps_to_plot) columns
fig, axes = plt.subplots(
    n_cycles_to_track, 
    len(steps_to_plot), 
    subplot_kw={'projection':'3d'}, 
    figsize=(18, 8)
)

# If axes is 1D, reshape to 2D for consistency
if n_cycles_to_track == 1:
    axes = np.expand_dims(axes, axis=0)
if len(steps_to_plot) == 1:
    axes = np.expand_dims(axes, axis=1)

# Loop over the selected time steps
for col_idx, step in enumerate(steps_to_plot):
    y0 = _ys[step]        # PH values or intensity
    A = Y_figure[step]    # 3D points of the trefoil
    max_c = y0.shape[1]   # number of cycles in this frame

    for row_idx, w in enumerate(what):
        ax = axes[row_idx, col_idx]

        if abs(w) > max_c:
            # If the cycle index does not exist, hide the axis
            ax.set_axis_off()
            continue

        # Plot normal background points in light gray and semi-transparent
        ax.scatter(
            A[:,1], A[:,0], A[:,2],
            s=40,
            color='lightgray',
            edgecolor='k',
            linewidth=0.5,
            alpha=0.3,
            rasterized=True
        )

        # Plot the selected cycle with size proportional to y0
        ax.scatter(
            A[:,1], A[:,0], A[:,2],
            s=5 + 100 * y0[:, w],
            edgecolor='k',
            linewidth=0.8,
            color=cycle_colors[row_idx % len(cycle_colors)],
            rasterized=True
        )

        # Hide axes for cleaner visualization
        ax.set_axis_off()

        # Set the 3D view angle
        ax.view_init(elev=50, azim=20, roll=0)

        # Set title above the first row
        if row_idx == 0:
            ax.set_title(f"Step {step}", fontsize=20)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
