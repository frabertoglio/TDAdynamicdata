```python
import numpy as np
import gudhi
import json
import csv
import matplotlib
import matplotlib.pyplot as plot
import math
import plotly.graph_objects as go
import os
import itertools
import import_ipynb
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
import glob
import re
```

```python
def plot_vines_3D(filename, num_vines=None):
    """
    Plot vines from a JSON filei

    Parameters
    ----------
    filename : str
        Name of JSON file with vines.
    num_vines : int or None, optional
        max number of vines to plot.  None = all.
    """
    full_path = os.path.join("vines", filename)
    with open(full_path, "r") as f:
        Vines = json.load(f)

    start_index = 2 if filename.endswith("0.json") else 1

    fig = go.Figure()
    all_b, all_d, all_t = [], [], []

    # number of vines to plot
    end_index = len(Vines) if num_vines is None else min(start_index + num_vines, len(Vines))

    for i in range(start_index, end_index):
        vine = Vines.get(f'Vine{i}')
        if vine is None:
            continue

        b = vine["births"]
        d = vine["deaths"]
        t = vine["vertex_times"]

        filtered = [(bi, di, ti) for bi, di, ti in zip(b, d, t) if bi != di]
        if not filtered:
            continue

        b_f, d_f, t_f = zip(*filtered)
        fig.add_trace(go.Scatter3d(
            x=b_f, y=d_f, z=t_f,
            mode='lines+markers',
            name=f'Vine {i}',
            line=dict(width=4)
        ))

        all_b.extend(b_f)
        all_d.extend(d_f)
        all_t.extend(t_f)

    # add plane y = x
    if all_b and all_t:
        min_val = min(min(all_b), min(all_d))
        max_val = max(max(all_b), max(all_d))
        max_t = max(all_t)

        xy_vals = np.linspace(min_val, max_val, 30)
        z_vals = np.linspace(0, max_t, 30)
        X, Z = np.meshgrid(xy_vals, z_vals)
        Y = X

        fig.add_trace(go.Surface(
            x=X, y=Y, z=Z,
            colorscale='Greys',
            opacity=0.4,
            showscale=False,
            name='Plane y = x'
        ))

    fig.update_layout(
        scene=dict(
            xaxis_title='Birth',
            yaxis_title='Death',
            zaxis_title='Time'
        ),
        title=f'Vines - {filename}',
        width=1000,
        height=1000
    )

    fig.show()
```

```python
def plot_vines_file(file, num_vines=None, inizio=1):
   
    base_path = os.path.join("vines/")
    """
   Plots the vines from a JSON file as a 3D vineyard plot.

Parameters
----------
file : str
    Path to the JSON file containing the vines.
num_vines : int or None, optional
    Maximum number of vines to plot. None = plot all.
start : int, optional
    Index of the first vine to include.
    """
    with open(base_path+file+".json", 'r') as f:
        Vines = json.load(f)

    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_title("Vineyard", fontsize=20)
    ax.set_xlabel("Birth")
    ax.set_ylabel("Death")
    ax.set_zlabel("T")

    all_b, all_d, all_t = [], [], []

    n_vines = len(Vines) if num_vines is None else min(num_vines + inizio, len(Vines))

    for i in range(inizio, n_vines):
        b = Vines['Vine' + str(i)]["births"]
        d = Vines['Vine' + str(i)]["deaths"]
        t = Vines['Vine' + str(i)]["vertex_times"]

        # Filter points with birth != death
        filtered_points = [(bi, di, ti) for bi, di, ti in zip(b, d, t) if bi != di]
        if not filtered_points:
            continue

        b_filtered, d_filtered, t_filtered = zip(*filtered_points)
        ax.plot(b_filtered, d_filtered, t_filtered, linewidth=1)
       # ax.scatter(b_filtered, d_filtered, t_filtered, s=5)

        all_b.extend(b_filtered)
        all_d.extend(d_filtered)
        all_t.extend(t_filtered)

    # Add plane y = x
    if all_b and all_d and all_t:
        min_val = min(min(all_b), min(all_d))
        max_val = max(max(all_b), max(all_d))
        max_t = max(all_t)
        x = np.linspace(min_val, max_val, 20)
        z = np.linspace(0, max_t, 20)
        X, Z = np.meshgrid(x, z)
        Y = X
        ax.plot_surface(X, Y, Z, alpha=0.2, color='gray')

    plt.tight_layout()

    plt.show()
```

```python
def plot_vines_file_side(file, num_vines=None, inizio=1):
    base_path = os.path.join("vines/")

    """
Plots the vines from a JSON file as a 3D vineyard plot "side by side",
showing two views of the same plot:
- left: standard view
- right: rotated view to highlight the y=x cutting plane

Parameters
----------
file : str
    Path to the JSON file containing the vines.
num_vines : int or None, optional
    Maximum number of vines to plot. None = plot all.
start : int, optional
    Index of the first vine to include.
    """
    
    with open(base_path + file + ".json", 'r') as f:
        Vines = json.load(f)

    fig = plt.figure(figsize=(20, 10))
    ax_left = fig.add_subplot(1, 2, 1, projection='3d')
    ax_right = fig.add_subplot(1, 2, 2, projection='3d')
    fig.suptitle("Linked Ellipses", fontsize=22)
    for ax in [ax_left, ax_right]:
        ax.set_xlabel("Birth", fontsize=16)
        ax.set_ylabel("Death" , fontsize=16)
        ax.set_zlabel("T" , fontsize=16)



    all_b, all_d, all_t = [], [], []

    n_vines = len(Vines) if num_vines is None else min(num_vines + inizio, len(Vines))

    for i in range(inizio, n_vines):
        vine = Vines.get('Vine' + str(i))
        if vine is None:
            continue

        b = vine["births"]
        d = vine["deaths"]
        t = vine["vertex_times"]

        filtered_points = [(bi, di, ti) for bi, di, ti in zip(b, d, t) if bi != di]
        if not filtered_points:
            continue

        b_f, d_f, t_f = zip(*filtered_points)
        for ax in [ax_left, ax_right]:
            ax.plot(b_f, d_f, t_f, linewidth=1)
        all_b.extend(b_f)
        all_d.extend(d_f)
        all_t.extend(t_f)

    # Piano y = x
    if all_b and all_d and all_t:
        min_val = min(min(all_b), min(all_d))
        max_val = max(max(all_b), max(all_d))
        max_t = max(all_t)
        x = np.linspace(min_val, max_val, 20)
        z = np.linspace(0, max_t, 20)
        X, Z = np.meshgrid(x, z)
        Y = X

        for ax in [ax_left, ax_right]:
            ax.plot_surface(X, Y, Z, alpha=0.2, color='gray')

    ##---Angles
    #ax_left.view_init(elev=30, azim=-60)   #  standard
    ax_right.view_init(elev=30,azim=50)   #  rotated

    plt.tight_layout()
    plt.show()
```
