```python
def compare_scale(X_ref, X_new):
    # Centroide e norma media rispetto al centro
    def spread(X):
        center = np.mean(X, axis=0)
        return np.mean(np.linalg.norm(X - center, axis=1))

    s_ref = spread(X_ref)
    s_new = spread(X_new)

    ratio = s_new / s_ref
    print(f"Scala X_ref: {s_ref:.4f}, X_new: {s_new:.4f}, Rapporto: {ratio:.4f}")

    return ratio

def rescale_to_match(X_new, X_ref):
    center_ref = np.mean(X_ref, axis=0)
    center_new = np.mean(X_new, axis=0)
    
    scale_ref = np.mean(np.linalg.norm(X_ref - center_ref, axis=1))
    scale_new = np.mean(np.linalg.norm(X_new - center_new, axis=1))
    
    scale_ratio = scale_ref / scale_new
    
    # Riscalamento e ricentraggio
    X_new_scaled = (X_new - center_new) * scale_ratio + center_ref
    return X_new_scaled      


```

```python
import numpy as np
import plotly.graph_objects as go
import os

def plot_3d_points(data, title="Plot 3D Interattivo", marker_color='blue', marker_size=5):
    """
    Plot interattivo 3D da nuvola di punti o file txt.

    Args:
        data (np.ndarray or str): matrice Nx3 di punti o path a file txt.
        title (str): titolo del grafico.
        marker_color (str): colore dei punti.
        marker_size (int): dimensione dei marker.

    Raises:
        ValueError: se l'input non è né array numpy né path valido.
    """
    if isinstance(data, str):
        if not os.path.isfile(data):
            raise ValueError(f"File {data} non trovato.")
        points = np.genfromtxt(data)
    elif isinstance(data, np.ndarray):
        points = data
    else:
        raise ValueError("Input deve essere un array numpy o il path a un file txt.")

    fig = go.Figure()
    fig.add_trace(go.Scatter3d(
        x=points[:,0],
        y=points[:,1],
        z=points[:,2],
        mode='markers',
        marker=dict(
            size=marker_size,
            color=marker_color,
            opacity=0.8
        ),
        name='Punti 3D'
    ))

    fig.update_layout(
        title=title,
        scene=dict(
            xaxis_title='X',
            yaxis_title='Y',
            zaxis_title='Z',
            aspectmode='data'
        ),
        width=800,
        height=600
    )

    fig.show()
```
