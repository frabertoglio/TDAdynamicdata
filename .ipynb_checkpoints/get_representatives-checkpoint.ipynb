```julia
using Ripserer
using DelimitedFiles, DataFrames, CSV
using JSON
using OrderedCollections
using Plots
```

```julia
return_points(representatives) = [Set(i for e in es for i in e) for es in representatives];
barcodes(PH, dim::Int) = hcat(collect.(collect(PH[dim+1]))...)';
representatives(PH, dim::Int) = [[collect(r.simplex) for r in collect(c)] for c in representative.(PH[dim+1])];
```

```julia
###----one step example
step = 20
name = "rings"
txt = true  

ext = txt ? ".txt" : ".tsv"

##-----file
if txt
    input_data = "simulations/" * name * "/" * name * string(step) * ext
    grid = Matrix{Float64}(CSV.read(input_data, DataFrame; header=false)) |> eachrow .|> Tuple;
else
    input_data = "simulations/" * name * "/step_" * string(step) * ext
    grid = Matrix{Float64}(CSV.read(input_data, DataFrame; delim='\t')) |> eachrow .|> Tuple;
end
```

```julia
scatter(grid; label="data", markersize=2, aspect_ratio=1)
```

```julia
PH = ripserer(grid; dim_max=1,  alg=:involuted) 
```

```julia
plot(PH, title="Persistence Diagram " * string(step) , markersize = 8,size = (400, 400))

#savefig(joinpath(output_dir, name * "_PD" * string(step) * ".pdf"))
#savefig("rings_PD"*string(step)*".pdf")
```

```julia
##------Series of persistence diagrams
name = "rings"
steps = 0:5:20
plots_array = []

for step in steps
    input_data = "simulations/" * name * "/" * name * "$step"*ext
    
    grid = Matrix{Float64}(CSV.read(input_data, DataFrame; header=false)) |> eachrow .|> Tuple
    PH = ripserer(grid; dim_max=1, alg=:involuted)
    
   
    p = plot(PH, markersize = 6,title="Persistence Diagram $step", xlabel="", ylabel="")
    push!(plots_array, p)
end

plt = plot(plots_array..., layout=(1,length(steps)), size=(300*length(steps), 300))
display(plt)
#savefig("rings_PH_5steps.pdf")
```

```julia
##------get representative cycles and barcodesJL for a given simulation
name="rings"
txt = true  

ext = txt ? ".txt" : ".tsv"

for step in 0:100
    if txt
        input_data = "simulations/" * name * "/" * name * "$step.txt"
        grid = Matrix{Float64}(CSV.read(input_data, DataFrame; header=false)) |> eachrow .|> Tuple;
    else
        input_data = "simulations/" * name * "/step_" * string(step) * ext
        grid = Matrix{Float64}(CSV.read(input_data, DataFrame; delim='\t')) |> eachrow .|> Tuple;
    end

    PH = ripserer(grid; dim_max=1, alg=:involuted)

    bar = barcodes(PH, 1)
    repre = representatives(PH, 1)
    repre = return_points(repre)

    ##---- read JSON file and add representatives, barcodes JL
    json_file = "PH/PH_" * name * "/PH_$step.json"
    dic = Dict{String,Any}()

    if isfile(json_file)
        dic = JSON.parsefile(json_file)
    end

    ordered_dic = OrderedDict(
    "barcode" => get(dic, "barcode", nothing),
    "persistence" => get(dic, "persistence", nothing),
    "barcodeJL" => bar,
    "representatives" => repre,
    "filtration" => get(dic, "filtration", nothing),
    )
    open(json_file, "w") do io
        JSON.print(io, ordered_dic, 2)
    end
end
```
